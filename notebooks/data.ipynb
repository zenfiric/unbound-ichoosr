## Input data

```python
import pandas as pd
# Read the CSV file
file_path = "../data/data/input_registration_sampledata_07032025.csv"
df = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print("Dataset Info:")
print(df.info())
```

```python
# Show first few rows
print("\nFirst few rows:")
df.head()
```

```python
# Display summary statistics
print("\nSummary Statistics:")
df.describe()
```

```python
# Check for missing values
print("\nMissing Values:")
df.isnull().sum().plot(kind='barh')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for better-looking graphs
plt.style.use('seaborn-v0_8')
fig = plt.figure(figsize=(15, 12))

# 1. Roof Type Distribution
plt.subplot(2, 2, 1)
sns.countplot(data=df, x='Product_RoofPitch')
plt.title('Distribution of Roof Types')
plt.xticks(rotation=45)

# 2. Roof Orientation Distribution
plt.subplot(2, 2, 2)
sns.countplot(data=df, x='Product_RoofOrientation')
plt.title('Distribution of Roof Orientations')
plt.xticks(rotation=45)

# 3. Home Size Distribution
plt.subplot(2, 2, 3)
sns.countplot(data=df, x='Product_HomeSize')
plt.title('Distribution of Home Sizes')
plt.xticks(rotation=45)

# 4. Number of Stories Distribution
plt.subplot(2, 2, 4)
sns.countplot(data=df, x='Product_NumberOfStories')
plt.title('Distribution of Number of Stories')

plt.tight_layout()
plt.show()

# Monthly Electricity Usage Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Product_MonthlyElectricityUsage', bins=30)
plt.title('Distribution of Monthly Electricity Usage')
plt.xlabel('Monthly Electricity Usage')
plt.show()

# Relationship between Home Size and Number of Stories
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Product_HomeSize', y='Product_NumberOfStories')
plt.title('Home Size vs Number of Stories')
plt.xticks(rotation=45)
plt.show()

# State-wise distribution (if there are multiple states)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Contact_state')
plt.title('Distribution by State')
plt.xticks(rotation=45)
plt.show()
```

## Output data

```python
import pandas as pd
# Read the CSV file
file_path = "../data/data/results_output_07032025.csv"
df = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print("Dataset Info:")
print(df.info())
```

```python
# Show first few rows
print("\nFirst few rows:")
df.head()
```

```python
# Display summary statistics
print("\nSummary Statistics:")
df.describe()
```

```python
# Check for missing values
print("\nMissing Values:")
df.isnull().sum().plot(kind='barh')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for better-looking graphs
plt.style.use('seaborn-v0_8')

# 1. System Size and Incentives Distribution
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# System Size Distribution
sns.histplot(data=df, x='Proposal_SystemSize', bins=20, ax=axes[0,0])
axes[0,0].set_title('Distribution of System Sizes')
axes[0,0].set_xlabel('System Size (kW)')

# Solar Panel Types
sns.countplot(data=df, x='OfferSolarPanelType', ax=axes[0,1])
axes[0,1].set_title('Distribution of Solar Panel Types')
axes[0,1].set_xlabel('OfferSolarPanel Type (W)')
axes[0,1].tick_params(axis='x', rotation=45)

# Total Incentives for LMI
sns.histplot(data=df, x='Proposal_TotalIncentivesForLmi', bins=20, ax=axes[1,0])
axes[1,0].set_title('Distribution of Total Incentives for LMI')
axes[1,0].set_xlabel('Incentive Amount ($)')

# Campaign Distribution
sns.countplot(data=df, x='campaign_name', ax=axes[1,1])
axes[1,1].set_title('Distribution by Campaign')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Relationship between System Size and Incentives
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Proposal_SystemSize', y='Proposal_TotalIncentivesForLmi')
plt.title('System Size vs Total Incentives for LMI')
plt.xlabel('System Size (kW)')
plt.ylabel('Incentives Amount ($)')
plt.show()

# Supplier Analysis
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Proposal_SupplierId', y='Proposal_SystemSize')
plt.title('System Sizes by Supplier')
plt.xticks(rotation=45)
plt.xlabel('Supplier ID')
plt.ylabel('System Size (kW)')
plt.show()
```
